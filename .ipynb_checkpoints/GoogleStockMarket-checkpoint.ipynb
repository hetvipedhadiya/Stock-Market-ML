```python
import pandas as pd
```

```python
data = pd.read_csv("GOOG.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Day_of_Week'] = data['Date'].dt.weekday  # Monday = 0, Sunday = 6
data['Is_Weekend'] = data['Day_of_Week'].apply(lambda x: 1 if x >= 5 else 0)  # 1 for Saturday/Sunday
data['Quarter'] = data['Date'].dt.quarter
```

```python
data = data.drop(columns=['Date'])
```

```python
data.head()
```

```python
x = data.drop("Volume",axis=1) #it means column wise row = 0
y = data["Volume"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
plt.scatter(y_pred,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)  # Lower is better
mse = mean_squared_error(y_test, y_pred)   # Lower is better
rmse = mean_squared_error(y_test, y_pred)  # Lower is better
r2 = r2_score(y_test, y_pred)  # Higher (closer to 1) is better

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_estimators=300, max_depth=20, min_samples_split=5, random_state=20)
```

```python
rf_model.fit(x_train, y_train)

# Predict
y_pred_rf = rf_model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_rf)
mse = mean_squared_error(y_test, y_pred_rf)
rmse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

print("Random Forest Results:")
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
plt.scatter(y_test,y_pred_rf)
plt.plot()
```
